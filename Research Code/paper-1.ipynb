```python
import numpy as np
import matplotlib.pyplot as plt
#import corner
```

```python
Y0,Y1=[],[]
z=[]
def target_1(z):
  num = (z*(7/8)*(1+z))/np.sqrt(0.1*pow(1+z,3)+0.9)
  deno = ((7/8)*(1+z))/np.sqrt(0.1*pow(1+z,3)+0.9)
  return num,deno

for i in np.arange(1,4,0.001):
  z.append(i)
  t0,t1 = target_1(i)
  Y0.append(t0)
  Y1.append(t1)

plt.figure(figsize=(9,7))
plt.plot(z, Y0)
plt.show()

plt.plot(z, Y1)
plt.show()
```

```python
from scipy.integrate import trapz

Y_=[]
z=np.arange(0,3,0.01)

for i in z:
  nu=(i*(7/8)*(1+i))/np.sqrt(0.1*pow(1+i,3)+0.9)
  x = np.linspace(0, i, 100)
  f=((7/8)*(1+x))/np.sqrt(0.1*pow(1+x,3)+0.9)
  I_trapz = trapz(f,x)
  Y_.append(nu/I_trapz)
```

```python
fig, ax = plt.subplots()
start, end = ax.get_xlim()
ax.yaxis.set_ticks(np.arange(0.8, 1.15,0.05))
ax.plot(z,Y_)
plt.show()
```

```python
#-----alpha-z relation------
from scipy.integrate import trapz

Y_=[]
z=np.arange(0,3,0.01)
OMG_M=[0.1,0.3,0.5]

for omg_M in OMG_M:
  tmp=[]
  for i in z:
    nu=(i*(7/8)*(1+i))/np.sqrt(omg_M*pow(1+i,3)+(1-omg_M))
    x = np.linspace(0, i, 100)
    f=((7/8)*(1+x))/np.sqrt(omg_M*pow(1+x,3)+(1-omg_M))
    I_trapz = trapz(f,x)
    tmp.append(nu/I_trapz)
  Y_.append(tmp)
```

```python
#-----Plot of alpha-z relation------
fig, ax = plt.subplots(figsize=(10,7))
ax.plot(z,Y_[0],label="Omg_m=0.1")
ax.plot(z,Y_[1],label="Omg_m=0.3")
ax.plot(z,Y_[2],label="Omg_m=0.5")
plt.xlabel('z',size='18')
plt.ylabel(r'$\alpha$',size='18')
plt.title(r'$\alpha$-z Relation',color='purple',size='20')
plt.legend()
plt.xlim()
plt.show()
```

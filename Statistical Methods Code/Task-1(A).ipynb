## #Transformations

```python
#----Tansformation equation for Standard Uniform Distribution----
a,b=3,2
X = np.random.uniform(0, 1, size=100000)
Y=a+(b-a)*X  # Transformation equation

plt.figure(figsize=(15,7))
plt.subplot(1,2,1)
plt.plot(X,Y,color='g')
plt.xlabel('x-axis ---> ',size='15')
plt.ylabel('y-axis ---> ',size='15')
plt.title('Transformation',color='purple',size='20')

plt.subplot(1,2,2)
plt.hist(Y,bins=20,edgecolor='black',color='c')
plt.xlabel('Values ---> ',size='15')
plt.ylabel('Frequency ---> ',size='15')
plt.title('Uniform Distribution',color='purple',size='20')

plt.show()
```

```python
#----Tansformation equation for Standard Normal Distribution---------------
# (USING SEABORN)------------
Z=np.random.normal(0,1,100000)
m=5
sd=2
N=m+sd*Z

plt.figure(figsize=(10,7))
sns.distplot(Z, bins=30, kde=True, color='darkmagenta')
sns.distplot(N, bins=30, kde=True, color='green')
plt.xlabel('Values ----> ',size='15')
plt.ylabel('Probablity Density ----> ',size='15')
plt.title('Normal Distribution',color='purple',size='20')
plt.show()
```

```python
# (USING MATPLOTLIB)-------------------
from scipy.stats import norm
import statistics

Z=np.random.normal(0,1,100000)
m1, s1 = 0,1

#---
plt.figure(figsize=(15,7))
plt.hist(Z, bins=30, edgecolor='black',density=1)
# Plot the PDF.
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, m1, s1)
#---
plt.plot(x, p, 'k', linewidth=2)
plt.title("Standard Normal Distribution",size='18')
plt.xlabel("Data Points",size='12')
plt.ylabel("Density",size='12')
plt.show()
```

```python
#------------------General Code for Normal Distribution---------------------------

#Mean and Standard Deviation
x=np.random.normal(0,1,100000)
x=np.sort(x)
m=np.mean(x)
sd=np.std(x)
#Probability Density Function
pdf=(np.exp(-0.5*((x-m)/sd)**2))/(sd*np.sqrt(2*np.pi))

#Plotting
plt.figure(figsize=(10,7))
plt.hist(x,bins=20,edgecolor='black',alpha=0.5,density="True")
plt.plot(x,pdf,color="red")
plt.title("Standard Normal Distribution")
plt.xlabel("Data Points")
plt.ylabel("Density")
plt.show()
```

```python
#To find the cost of Pudding using Monte Carlo Simulation
import pandas as pd
N=1                     
mean=np.array([]) 

while(N<=1000):
  cost=np.array([])
  milk= np.random.uniform(18,20,N)          #for one lt
  sugar= np.random.uniform(20,23,N)           #for one kg
  bread = np.random.normal(loc=25,scale=1,size=N)  #for 12 slices
  egg= np.random.choice([2,2.5,3],p=[0.2,0.5,0.3],size=N) #For 1 egg
  almond= np.random.choice([500,700],p=[0.7,0.3],size=N) #For 0.5 kg

  for i in range(0,N):
    cost_per=(milk[i] + 2*egg[i] + 0.1*sugar[i] + (1/6)*bread[i] + 0.1*almond[i])    
    cost=np.append(cost,cost_per)
  
  #print('For value of N:',N)
  mean_value=sum(cost)/N     #To find the mean cost for particular data points generated
  mean=np.append(mean,mean_value) #Append the mean_value to mean ndarray
  #print('Mean of cost of pudding:',mean_value)
  #plt.hist(cost,bins=100)   #To plot the histogram 
  #plt.show()
  
  N+=1    #To increment the number of points generated
#plt.hist(cost,bins=100)   #To plot the histogram 
  #plt.show()

d=np.array([cost,almond,milk,sugar,bread,egg]).T
#print(d)
print('Mean of cost of pudding:',mean_value,"\n")
data=pd.DataFrame(data=d, columns=["cost","almond","milk","sugar","bread","egg"])
plt.hist(cost,bins=100)   #To plot the histogram 
plt.show()
```

```python
# Importing Libraries 
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import seaborn as sn
import math
```

## **Value of Pi (Circle inscribed in a square)**

```python
#Plotting a circle inside the square
r=1
angle=np.linspace(0,np.pi*2)
x1=r*np.cos(angle)
y1=r*np.sin(angle)

plt.figure(figsize=(6,6))
plt.plot(x1,y1,color="purple",markersize=1)
plt.xlim(-r,r)
plt.ylim(-r,r)
plt.xlabel('----x axis----',size='12')
plt.ylabel('----y axis----',size='12')
plt.show()
```

## **Monte Carlo Simulation**

```python
#--------GETTING RANDOM VARIABLES----------
N=int(input("Enter the value of N = ")) #Total Number of turns
r=int(input("Radius of the circle = "))
c=0  #times when target hits inside the circle
x=np.random.uniform(0,r,N)  #Syntax: np.random.uniform(low,high,size)
y=np.random.uniform(0,r,N)
N1=np.linspace(1,N,N)

#-------------Empty List-----------------
f=[]
pi_o=[]

for i in range(N):
  if (x[i]**2+y[i]**2)<=r**2:
    c=c+1
  pi=4*c/(i+1)
  pi_o.append(3.14)
  f.append(pi)
print("Value of pi using Monte Carlo Simulation = ",pi)

#-------------Plotting---------------

plt.figure(figsize=(20,7))
plt.subplot(1,2,1)      #Row 1 Column 2 Index 1
plt.scatter(x,y,color="red",marker=".",s=10)
angle=np.linspace(0,np.pi/2,150)
x1=r*np.cos(angle)
y1=r*np.sin(angle)
plt.plot(x1,y1,color="blue",marker="o",markersize=1)
plt.xlabel('----x axis----',size='20',color='brown')
plt.ylabel('----y axis----',size='20',color='brown')
plt.title("Scattered plot",size='30',color='purple')

plt.subplot(1,2,2)
plt.plot(N1,f,color="brown",)
plt.plot(N1,pi_o,linestyle="dashed",color="green")
plt.xlim(0,N)
plt.xlabel("---No. of turns---",size='20',color='brown')
plt.ylabel("---Pi---",size='20',color='brown') 
plt.title("Monte Carlo Simulation",size='30',color='purple')
plt.grid()
plt.show()
```

##**Value of pi (Sphere inscribed in a cube)**

```python
#-----Generate points------
N=int(input("Enter the value of N: "))
r=int(input("Radius of the sphere = "))
x,y,z= np.random.uniform(-r,r,size=N), np.random.uniform(-r,r,size=N), np.random.uniform(-r,r,size=N);
c=0      #Count for no. of times dart is hit inside the sphere

#-------------Empty List-----------------
N1=np.linspace(1,N,N)
f=[]
pi_o=[]

for i in range(N):
    if (x[i]**2 + y[i]**2 + z[i]**2) <=r**2 :
      c=c+1
    pi=6*c/(i+1)
    f.append(pi)
    pi_o.append(3.14)
print("Value of pi using Monte Carlo Simulation = ",pi)

#-------PLOTTING---------
plt.figure(figsize=(10,10))
ax=plt.axes(projection='3d')  #Syntax for 3d projection

#Adding x and y gridlines
ax.grid(color ='grey',linestyle ='-.', linewidth = 0.3,alpha = 0.2)
ax.scatter3D(x,y,z,alpha=0.7,marker='.',color='purple')
plt.title("3D scatter plot",fontsize='25',color='purple')

#Points for mesh Grid
u, v = np.mgrid[0:2*np.pi:200j, 0:np.pi:100j]

#Setting x,y,z Co-ordinates 
x1=r*np.cos(u)*np.sin(v)
y1=r*np.sin(u)*np.sin(v)
z1=r*np.cos(v)

ax.plot_wireframe(x1, y1, z1, rstride = 5, cstride = 5, linewidth = 1,color='red')
 
ax.set_xlabel('X-axis', fontweight ='bold',fontsize='15',color='brown')
ax.set_ylabel('Y-axis', fontweight ='bold',fontsize='15',color='brown')
ax.set_zlabel('Z-axis', fontweight ='bold',fontsize='15',color='brown')
plt.show()

plt.figure(figsize=(7,7))
plt.plot(N1,f,color="brown",)
plt.plot(N1,pi_o,linestyle="dashed",color="green")
plt.xlim(0,N)
plt.xlabel("---No. of turns---",size='20',color='brown')
plt.ylabel("---Pi---",size='20',color='brown')
plt.title("Monte Carlo Simulation",size='30',color='purple')
plt.grid()
plt.show()
```

# **Integration**

### `1) Univariate`

```python
#-------------Method-1--------------------------

N=100000
#----------Integral between 0 to pi-----------------
x=np.random.uniform(0,np.pi,N)
y=np.random.uniform(-1,1,N)
y1=[]
x.sort()

plt.figure(figsize=(10,7))
plt.scatter(x,y,s=5)

for i in range(len(x)):
  y1.append(math.sin(x[i]))
plt.plot(x,y1,color='red')

#----------
c=0
for i,j in zip(y,y1):
  if i<=j and i>=0:
    c+=1
I0=2*(np.pi-0)*c/N

#-----------Integral between pi to 2 pi---------------
x=np.random.uniform(np.pi,np.pi*2,N)
y=np.random.uniform(-1,1,N)
y2=[]
x.sort()

plt.figure(figsize=(10,7))
plt.scatter(x,y,s=10)

for i in range(len(x)):
  y2.append(math.sin(x[i]))
plt.plot(x,y2,color='red')
plt.show()

#----------
c1=0
for i,j in zip(y,y2):
  if i>=j and i<=0:
    c1+=1
I1=2*(np.pi*2-np.pi)*c1/N

#----------
I=I0+I1
print('\nIntegral= ',I)
```

```python
#---------------------Method-----2-----------------
import math

N=10000
x=np.random.uniform(0,np.pi*2,N)
y=np.random.uniform(-1,1,N)
y1=[]

plt.figure(figsize=(10,7))
x.sort()
for i in range(len(x)):
  y0=math.sin(x[i])
  if x[i]<=np.pi:
    if y[i]>=0 and y[i]<=y0:
      plt.scatter(x[i],y[i],s=5,alpha=0.6,c='cadetblue')
    else:
      plt.scatter(x[i],y[i],s=5,alpha=0.6,c='y')
  else:
    if y[i]<=0 and y[i]>=y0:
      plt.scatter(x[i],y[i],s=5,alpha=0.6,c='cadetblue')
    else:
      plt.scatter(x[i],y[i],s=5,alpha=0.6,c='y')
  y1.append(y0)
  
plt.plot(x,y1,color='red')
plt.axhline(y=0,linestyle='--',linewidth=2.5,c='black')
plt.xlabel("---x-axis---",size='20',color='brown')
plt.ylabel("---y-axis---",size='20',color='brown')
plt.title("Contour to be Integrated",size='30',color='purple')
plt.grid()
plt.show()
```

```python
#---------------------------WITHOUT TOLERANCE------------------------------

N=10000
#----------Integral between 0 to pi-----------------
x=np.random.uniform(0,2*np.pi,N)
y=np.random.uniform(-1,1,N)
y1=[]
x.sort()
c=0

plt.figure(figsize=(10,7))
for i in range(len(x)):
  y0=math.sin(x[i])
  if x[i]<=np.pi:
    if y[i]>=0 and y[i]<=y0:
      plt.scatter(x[i],y[i],s=5,alpha=0.6,c='cadetblue')
      c+=1
    else:
      plt.scatter(x[i],y[i],s=5,alpha=0.6,c='y')

  else:
    if y[i]<=0 and y[i]>=y0:
      plt.scatter(x[i],y[i],s=5,alpha=0.6,c='cadetblue')
      c+=1
    else:
      plt.scatter(x[i],y[i],s=5,alpha=0.6,c='y')

  y1.append(y0)

print('Total no. of darts inside the contour = ',c)
I=2*(2*np.pi-0)*c/N  
print("\nIntegral with 3 decimal place accuracy: ",I)
plt.plot(x,y1,color='red')
plt.axhline(y=0,linestyle='--',linewidth=2.5,c='black')
plt.xlabel("---x-axis---",size='20',color='brown')
plt.ylabel("---y-axis---",size='20',color='brown')
plt.title("Contour to be Integrated",size='30',color='purple')
plt.grid()
plt.show()      
```

```python
#----------------------------------WITH TOLERANCE----------------------------------
import math
import sympy as sy
from scipy.integrate import quad

N=100000
e=1
IN=0

#plt.figure(figsize=(10,7))
# To get Integral of given function in given interval with 3 decimal place accuracy-------
while e>0.0005:
  N+=1000
  x=np.random.uniform(0,2*np.pi,N)
  y=np.random.uniform(-1,1,N)
  y1=[]
  x.sort()
  c=0

  for i in range(len(x)):
    y0=math.sin(x[i])
    if x[i]<=np.pi:
      if y[i]>=0 and y[i]<=y0:
        #plt.scatter(x[i],y[i],s=5,alpha=0.6,c='cadetblue')
        c=c+1
    else:
      if y[i]<=0 and y[i]>=y0:
        #plt.scatter(x[i],y[i],s=5,alpha=0.6,c='cadetblue')
        c=c+1

    y1.append(y0)

  I=2*(2*np.pi-0)*c/N         #2 is multiplied because y is ranging from -1 to 1
  #print(I)
  e=abs(I-IN)
  IN=I

print("\nIntegral with 3 decimal place accuracy: ",IN)
```

### `2) Multivariate`

```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D  
# Axes3D import has side effects, it enables using projection='3d' in add_subplot

import matplotlib.pyplot as plt
import random

V=1*1*1 # Target Hyper volume
def fun(x, y):
    return y*x**2

N=10000
fig = plt.figure(figsize=(15,9))
ax = fig.add_subplot(111, projection='3d')
x = np.random.uniform(0,1,N)
y = np.random.uniform(0,1,N)
z = np.random.uniform(0,1,N)
Zf = fun(x,y)

ax.scatter3D(x, y, z, color = "yellow", marker='.')

X, Y = np.meshgrid(x, y)
tmp = np.array(fun(np.ravel(X), np.ravel(Y)))
Z = tmp.reshape(X.shape)

ax.plot_surface(X, Y, Z)

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
plt.show()

#------
c1=0
for i,j in zip(z,Zf):
    if i<=j:
      c1+=1

I=V*c1/N
print('\nIntegral= ',I)
```
